## Magnetocaloric effect

```python
diff_temperature = np.diff (Temperature)
diff_magnetization = np.diff (Magnetizaion[1:,:])
diff_field = np.diff (Field)

MCE = (diff_field[:,np.newaxis]*diff_magnetization)/diff_temperature


Field_index = (11,16,21,23,25,27,29)
sum_MCE = []
sum_field = []
for i in (Field_index):
    sum_MCE.append(-np.sum(MCE[:i-1,:], axis=0))
    sum_field.append (np.sum(diff_field[:i-1]))



Temp_ave = (Temperature[1:] + Temperature[:-1]) /2

fig = plt.figure (figsize=(8,6),dpi=100)
axes1 = fig.add_axes ([0.13,0.12,0.85,0.85])
NUM_COLORS = 7
markers = ['o', 'X', '>', 'D','v', 'P','s']
cm = plt.get_cmap('gnuplot')
axes1.set_prop_cycle('color', [cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])
for i,m in zip(range (0, len(sum_MCE)), markers):
    axes1.plot (Temp_ave, sum_MCE[i],marker = m, markersize=6, linewidth=0.5, label = f'{sum_field[i]:.0f} T')
axes1.set_xlabel ('$\mathrm{T~(K)}$', size=20)
axes1.set_ylabel (r'$\mathrm{-\Delta S_{m}~(J/kg~K)}$', size=20)
axes1.legend(prop={'size': 12}, loc='upper left', frameon=False)
axes1.set_xlim(50,370)
axes1.set_ylim(0,2)

axes1.tick_params(which='both', labelsize=16, width=1.5)
axes1.tick_params(which='major', length=5)
axes1.tick_params(which='minor', length=3)
axes1.set_yticks (np.arange(0,2.5,0.5))
axes1.text(340,1.8, "(a)", color = 'k', fontsize = 20)
axes1.xaxis.set_minor_locator(AutoMinorLocator(2))
axes1.yaxis.set_minor_locator(AutoMinorLocator(2))
for axis in ['top','bottom','left','right']:
    axes1.spines[axis].set_linewidth(1.5)


# plt.savefig('B_LSMO_MCE')
plt.show()
```
